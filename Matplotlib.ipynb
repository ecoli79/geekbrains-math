##### Задание 4*

```python
credit_cards = pd.read_csv('./creditcard.csv')
class_counts = credit_cards['Class'].value_counts()
class_counts # проверим сколько получается наблюдей для каждого значения Class 
percent_frauds = (class_counts.values[1] / class_counts.values[0]) * 100
print(f"Количество мошеннических транзакций составляет: {round(percent_frauds,3)}%")
```

```python
# стоим столбчатую диаграмму с распределением мошенических и остальных транзакций. На этой диаграмме
# мошенические транзакции теряются, из-за большой разницы в итоговых значениях
plt.style.use('fivethirtyeight')
class_counts.plot(kind="bar")
plt.show()
```

```python
# применима логарифмирование к полученных выше данным
class_counts.plot(kind="bar", logy=True)
plt.show()
```

```python
fraud = credit_cards.Class == 1
normal = credit_cards.Class == 0
data_fraud = credit_cards['V1'][fraud]
data_normal = credit_cards['V1'][normal]
plt.hist(data_fraud, bins=20, alpha=0.5, color='red', label='Class 1')
plt.hist(data_normal, bins=20, alpha=0.5, color='gray', label='Class 0')
plt.yscale('log')
plt.xlabel('V1')
plt.legend()
plt.show()
```

```python
#Создать одномерный массив а, состоящий из последовательности целых чисел от 0 до 12.
#Поменять форму этого массива, чтобы получилась матрица A (двумерный массив Numpy),
#состоящая из 4 строк и 3 столбцов. Получить матрицу At путем транспонирования матрицы A.
#Получить матрицу B, умножив матрицу A на матрицу At с помощью матричного умножения.
#Какой размер имеет матрица B? Получится ли вычислить обратную матрицу для матрицы B и почему?
a = np.arange(0,12,1)
A = np.reshape(a, (4,3))
At = A.T
B = np.dot(A,At)
det = np.linalg.det(B)
det # так как определитель матрицы B равен нулю, то обратную матрицу B не найти
print(f"Матрица А:\n{A}")
print("-"*48)
print(f"Транспонированная матрица At:\n{At}")
print("-"*48)
print(f"Матрица B:\n{B} \nи ее определитель:{det}")
```

```python
#Инициализируйте генератор случайных числе с помощью объекта seed, равного 42
np.random.seed(42)
c = np.random.randint(0,16,16)
C = np.reshape(c,(4,4))
D = (B*10) + C
D_det = np.linalg.det(D)
D_rank = np.linalg.matrix_rank(D)
D_inv = np.linalg.inv(D)
print(f"Ранк матрицы D равен:\n{D_rank}")
print("-"*48)
print(f"Определитель матрицы D равен: {D_det}")
print("-"*48)
print(f"Обратная матрица от матрицы D:\n{D_inv}")
```

```python
#Прировнять к 0 отрицательные числа в матрице D_inv а положительные к единице
D_inv = np.where(D_inv<0, 0,1)
D_inv
E = np.where(D_inv==1, B,C)
print(B)
print(C)
print(D_inv)
print(E)
```
