#Lesson 3 example 3-2

Узнать принадлежит ли прямая $$ \frac{x - x_1}{x_2 - x_1} = \frac{y - y_1}{y_2 - y_1} =\frac{z - z_1}{z_2 - z_1}$$
плоскости
$$A_1x + B_1y + C_1z + D_1 = 0$$
Для того, чтобы получить ответ на этот вопрос необходимо решить систему уравнений:

$$ 
\begin{equation*}
 \begin{cases}
   A_1(x - x_1) + B_1(y - y_1) + C_1(z - z_1) + D_1 = 0 
   \\
   A_1(x_2 - x_1) + B_1(y_2 - y_1) + C_1(z_2 - z_1) + D_1=0
 \end{cases}
\end{equation*}
$$
