```python
#Lesson-3 example 3
fig, axs = plt.subplots(2,2)
#circle
# it is radius
R = 4
# number of points
n = 50
# calclutating edges
t = np.linspace(0, 2*np.pi, n + 1)
x_c = R*np.cos(t)
y_c = R*np.sin(t)
axs[0, 0].plot(x_c,y_c)
axs[0, 0].axis('equal')
axs[0, 0].set_title('circle', fontsize = 10)
#ellipse
x_e = 10*np.cos(t)
y_e = 1*np.sin(t)
axs[0, 1].plot(x_e,y_c)
axs[0, 1].axis('equal')
axs[0, 1].set_title('ellipse', fontsize = 10)
#hyperbole
a = 2
b = 1.5
x_h = np.linspace(-100,100)
y_h = np.sqrt((x_h**2*b**2)/a**2 - 1*b**2)
axs[1, 0].plot(x_h,y_h)
axs[1, 0].plot(x_h,-y_h)
axs[1, 0].axis('equal')
axs[1, 0].set_title('hyperbole', fontsize = 10)
```

```python
#Lesson-3 example 5
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(131,projection='3d')
x = np.linspace(20, 40, 100)
y = np.linspace(20, 40, 100)
X, Y = np.meshgrid(x, y)
Z = 2*X + 2*Y
Z2 = 2*x + 2*Y + 100
ax.contour3D(X, Y, Z,50,cmap='binary')
ax.contour3D(X, Y, Z2,50,cmap='binary')
ax.view_init(30,20)


ax = fig.add_subplot(132,projection='3d')
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
x = 10 * np.outer(np.cos(u), np.sin(v))
y = 10 * np.outer(np.sin(u), np.sin(v))
z = 10 * np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_surface(x, y, z, rstride=4, cstride=4)

ax = fig.add_subplot(133,projection='3d')
x = np.linspace(-1, 1, 100)
z = np.linspace(-2, 2, 100)
Xc, Zc = np.meshgrid(x, z)
Yc = np.sqrt(1 - Xc**2)
rstride = 20
cstride = 10
ax.plot_surface(Xc, Yc, Zc, rstride=rstride, cstride=cstride)
ax.plot_surface(Xc, -Yc, Zc, rstride=rstride, cstride=cstride)


```

```python
#Lesson-3 example3-4
x = np.linspace(-3, 3, 200)
plt.grid()
plt.plot(x, (x**2 -1))
plt.plot(x, (np.exp(x) + x - 1)/x)

def equ(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) + x - x*y - 1)
p = (2, 2)
x1, y1 = fsolve(equ, p)
print(x1, y1) 

def equ2(p):
    x, y = p
    if (np.exp(x) + x(1 - y)) > 1:
        return x**2 - 1

p = (1, 1)
x1, y1 = fsolve(equ2, p)
print(y1)
```
